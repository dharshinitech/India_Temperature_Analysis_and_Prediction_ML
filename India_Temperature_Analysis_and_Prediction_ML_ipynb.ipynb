```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Load the data properly
# low_memory=False is added because your dataset has mixed types
data = pd.read_csv('city_temperature.csv', low_memory=False)

# 2. Filtering with the exact names found in your dataset list
chennai_data = data[data['City'] == 'Chennai (Madras)'].copy()
delhi_data = data[data['City'] == 'Delhi'].copy()

# 3. Data Cleaning: Removing -99 (invalid) temperature values
chennai_data = chennai_data[chennai_data['AvgTemperature'] != -99]
delhi_data = delhi_data[delhi_data['AvgTemperature'] != -99]

# 4. Yearly Grouping (to make the graph lines smooth)
chn_yearly = chennai_data.groupby('Year')['AvgTemperature'].mean()
del_yearly = delhi_data.groupby('Year')['AvgTemperature'].mean()

# 5. Plotting the Graph
plt.figure(figsize=(12, 6))
plt.plot(chn_yearly.index, chn_yearly.values, label='Chennai (Madras)', color='blue', marker='o')
plt.plot(del_yearly.index, del_yearly.values, label='Delhi', color='green', marker='s')

plt.xlabel("Year")
plt.ylabel("Average Temperature (°C)")
plt.title("Climate Trend Analysis: Chennai vs Delhi")
plt.legend()
plt.grid(True)
plt.show()

# 6. ML Prediction for Chennai 2030
X = chennai_data[['Year']]
y = chennai_data['AvgTemperature']
model = LinearRegression()
model.fit(X, y)
pred_2030 = model.predict([[2030]])
print(f"\nPredicted Temperature for Chennai in 2030: {pred_2030[0]:.2f} °C")
```
